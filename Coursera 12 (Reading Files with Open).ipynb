###### Better Way to Open a File

###### Using the with statement is better practice, it automatically closes the file 
even if the code encounters an exception. The code will run everything in the indent block then close the file object.

```python
with open(r"C:\Users\User\Documents\Python Scripts\Week 4\Data Set\countries of the world.txt")as File1:
    FileContent = File1.read()
    print(FileContent)
```

```python
# Verify if the file is closed

File1.close()
```

```python
# See the content of file  (As written in the with code above)

print(FileContent)
```

```python
# Read one line

with open(r"C:\Users\User\Documents\Python Scripts\Week 4\Data Set\countries of the world.txt")as File1:
    print('First Line =', File1.readline())
    print('Second Line =', File1.readline())
```

###### We can use a loop to iterate through each line:

```python
# Iterate through the lines (Iterate means Repeat a process)

with open(r"C:\Users\User\Documents\Python Scripts\Week 4\Data Set\countries of the world.txt")as File1:
    i = 0
    for line in File1:
        print(i,line)        # i = index    line does the same as x did in for loop.
        i = i+1              # This line prints the current iteration number (i), along with the line content. 
    
```

###### We can use the method readlines() to save the text file to a list:

```python
# Read all lines and save as a list

with open(r"C:\Users\User\Documents\Python Scripts\Week 4\Data Set\countries of the world.txt")as File1:
    Readlines = File1.readlines()
```
